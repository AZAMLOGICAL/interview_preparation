```python
# Given a DataFrame of sales, calculate the total revenue per region and product category.
```

```python
import pandas as pd

# Sample data
data = {
    'region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'product_category': ['Electronics', 'Furniture', 'Electronics', 'Furniture', 
                         'Furniture', 'Electronics', 'Furniture', 'Electronics'],
    'units_sold': [100, 50, 120, 80, 60, 90, 70, 110],
    'unit_price': [300, 150, 320, 200, 160, 310, 180, 330]
}

df = pd.DataFrame(data)
df['sales'] = df['units_sold'] * df['unit_price']
df.groupby(['region', 'product_category'])['sales'].sum().reset_index()
```

```python
df.groupby(['region', 'product_category']).agg(
    total_sales = ('sales', 'sum'),
    mean_sales = ('sales', 'mean')
).reset_index()
```

```python
# Q: Given a DataFrame with daily sales data, how would you resample it to get monthly total sales?
```

```python
# # Create a date range for daily data
date_range = pd.date_range(start='2024-01-01', end='2024-03-31', freq='D')

# Create a sample DataFrame with random sales values
df = pd.DataFrame({
    'date': date_range,
    'sales': np.random.randint(100, 500, size=len(date_range))
})

# Set 'date' as the index (important for resampling)
df.set_index('date', inplace=True)
```

```python
DF = df.resample('M').sum()
DF
```

```python
# Q: Create a pivot table showing average sales by region and product_category.
```

```python
import pandas as pd

# Sample data
data = {
    'region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'product_category': ['Electronics', 'Furniture', 'Electronics', 'Furniture',
                         'Clothing', 'Clothing', 'Furniture', 'Electronics'],
    'sales': [200, 150, 300, 250, 180, 120, 230, 310]
}

df = pd.DataFrame(data)
df
```

```python
pd.pivot_table(index='region', columns='product_category', values='sales', data=df, aggfunc='mean')
```

```python
pd.pivot(index='region', columns='product_category', values='sales', data=df)
```

```python
# Q: How do you rank products by sales within each region?
import pandas as pd

# Sample data
data = {
    'region': ['North', 'North', 'South', 'South', 'East', 'East', 'West', 'West'],
    'product': ['TV', 'Laptop', 'TV', 'Laptop', 'Phone', 'Tablet', 'Phone', 'Laptop'],
    'sales': [300, 450, 500, 400, 200, 250, 350, 600]
}

df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
df_region_mean = df.groupby('region')['sales'].mean().reset_index()
df_region_mean['rank'] = df_region_mean['sales'].rank()
df_region_mean
```

```python
# Q: Replace missing values in the discount column with 0 only if the status is 'completed'
```

```python
import pandas as pd
import numpy as np

# Sample data
data = {
    'order_id': [101, 102, 103, 104, 105],
    'status': ['completed', 'pending', 'completed', 'cancelled', 'completed'],
    'discount': [10, np.nan, np.nan, 5, np.nan]
}

df = pd.DataFrame(data)

# Display the DataFrame
df.loc[(df['status'] == 'completed') & (df['discount'].isna()), 'discount'] = 0
```

```python
df
```

```python
# Q: Given a column of lists (e.g. tags or categories), how do you expand them into individual rows using explode()?
```
